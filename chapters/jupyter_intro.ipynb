# Task
Analyze the relationship between age, marital status, and earner status across different religions using descriptive statistics and visualizations.

## Data preparation

### Subtask:
Clean and filter the data to include only the relevant columns ('Age', 'Marital_status', 'Earner', 'Religion') and drop rows with missing values in these columns.


**Reasoning**:
Create a new DataFrame with the specified columns and then drop rows with any missing values in these columns. Finally, display the head of the cleaned DataFrame.



```python
analysis_df = df[['Age', 'Marital_status', 'Earner', 'Religion']].copy()
analysis_df.dropna(inplace=True)
display(analysis_df.head())
```

## Descriptive statistics

### Subtask:
Calculate descriptive statistics (mean, median, standard deviation) for 'Age' grouped by 'Religion', 'Marital_status', and 'Earner' status.


**Reasoning**:
Calculate descriptive statistics for Age grouped by Religion and Marital_status, and then by Religion and Earner status.



```python
age_marital_religion_stats = analysis_df.groupby(['Religion', 'Marital_status'])['Age'].agg(['mean', 'median', 'std'])
age_earner_religion_stats = analysis_df.groupby(['Religion', 'Earner'])['Age'].agg(['mean', 'median', 'std'])
display(age_marital_religion_stats)
display(age_earner_religion_stats)
```

## Visualization

### Subtask:
Create visualizations such as boxplots, bar charts, or scatterplots to show the relationship between age, marital status, earner status, and religion.


**Reasoning**:
Create boxplots to visualize the distribution of Age across different Marital Statuses and Earner statuses, color-coded by Religion, as requested in the instructions.



```python
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.boxplot(x='Marital_status', y='Age', hue='Religion', data=analysis_df)
plt.title('Age Distribution by Marital Status and Religion')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Earner', y='Age', hue='Religion', data=analysis_df)
plt.title('Age Distribution by Earner Status and Religion')
plt.tight_layout()
plt.show()
```
